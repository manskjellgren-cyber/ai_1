### Select the rows where the column "company" is equal to 'honda'. 

```python
import numpy as np
import pandas as pd
cars = pd.read_csv("cars_data.csv")
# From cars dataframe, chose column company and if a row in company contains honda
# place it in a filtered spreadataframe called hondas
hondas = cars[cars['company'].str.contains('honda')]
hondas
```

### Select the rows where the column "company" is equal to any of the values (audi, bmw, porsche).

```python


import numpy as np
import pandas as pd
cars = pd.read_csv("cars_data.csv")
# if a row in company contains audi or | bmw or | porsche
# place it in a dataframe called germans
germans = cars[cars['company'].str.contains('audi') | cars['company'].str.contains('bmw') | cars['company'].str.contains('porsche')]
germans

```

### Find the number of cars (rows) for each company. 

```python
import numpy as np
import pandas as pd
cars = pd.read_csv("cars_data.csv")
# Skapar en kolumn för varje unikt ord (separerat med mellanslag)
car_volume= cars['company'].str.get_dummies(sep=' ')
# Summan av varje kolumn ger totala förekomsten av det ordet
print(car_volume.sum())
```

### Find the maximum price for each company. 

```python
import numpy as np
import pandas as pd
cars = pd.read_csv("cars_data.csv")
# Find index for the row with highest price per company
maxindex = cars.groupby('company')['price'].idxmax()
# Fetch the rows with these index
maxprice = cars.loc[maxindex]
print(maxprice) 
```

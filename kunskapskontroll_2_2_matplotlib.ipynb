### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
import numpy as np
import matplotlib.pyplot as plt

print("Plotting in matplotlib can be done")
print("as Object Oriented style or pyplot style")
print("The recommended approach is the Object Oriented approach")
#the OO style

print("\nThe Object Oriented approach (first figure)")
x = np.linspace(0, 2, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x, label='linear')  # Plot some data on the Axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the Axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the Axes.
ax.set_ylabel('y label')  # Add a y-label to the Axes.
ax.set_title("Simple Plot")  # Add a title to the Axes.
ax.legend()  # Add a legend.

#the pyplot style

print("The Pyplot approach (second figure)")
x = np.linspace(0, 2, 100)  # Sample data.

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) Axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
print("A Figure equals the complete possible area for data to be visualized.")
print("An Artist. Basically, everything visible on the figure is an Artist, like the figure, Axes and Axis objects.")
print("Most Artists are tied to an Axes. Such an artist cannot be shared by, or moved between, multiple Axes")
print("When the figure is rendered, all of the artists are drawn to the canvas.")
print("Axis set the scale and limits and generate ticks on x-y-z metrics on 1D, 2D and 3D visualizations")
print("Axes is an artist attached to a figure that contains a region for plotting data")
print("Axes are the primary interface for configuring most parts of your plot:")
print("Adding data, controlling axis, scales and limits, adding labels etc.")
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4., 4., 100)  # Sample data. (from -4 to 4 with 100 samples)
# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(4, 4), layout='constrained')
ax.plot(x, x**2, label='y=x^2')  # Plot more data on the Axes...
ax.set_xlabel('x label')  # Add an x-label to the Axes.
ax.set_ylabel('y label')  # Add a y-label to the Axes.
ax.set_title("Object Oriented Approach (EXPLICIT)")  # Add a title to the Axes.
ax.legend()  # Add a legend.

#the pyplot style

x = np.linspace(-4.,4., 100)  # Sample data.
plt.figure(figsize=(4, 4), layout='constrained')
# Plot some data on the (implicit) Axes.
plt.plot(x, x**2, label='y=x^2')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Pyplot approach (IMPLICIT)")
plt.legend()
```
